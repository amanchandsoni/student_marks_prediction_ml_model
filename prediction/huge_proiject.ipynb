```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv("Final_Marks_Data.csv")

X = data[['Daily Study Hours',
          
        'Attendance (%)',
        'Internal Test 1 (out of 40)',
        'Assignment Score (out of 10)',
        'Internal Test 2 (out of 40)'
        ]]
Y = data['Final Exam Marks (out of 100)']

model = LinearRegression()

model.fit(X, Y)
predicted_score = model.predict(X)

mae = mean_absolute_error(Y, predicted_score )
mse = mean_squared_error(Y, predicted_score)
rmse = np.sqrt(mse)
r2_score = r2_score(Y, predicted_score)

print("mean_absolute_error",round(mae, 2))

print("mean_squared_error",round(mse, 2))
print("root_mean_absolute_error",round(rmse, 2))
print("r2 score is :",round(r2_score, 2))


plt.figure(figsize=((10, 6)))
plt.hist(data["Final Exam Marks (out of 100)"], bins=30, color='skyblue', edgecolor='black')
plt.title("distribution of marks")
plt.xlabel("final exam score")
plt.ylabel("number of students")
plt.grid = True
plt.show()


plt.figure(figsize=((10, 6)))
plt.scatter(data['Daily Study Hours'], Y,  color='blue', label='actual scores')
plt.plot(X, predicted_score, color="red", label = 'predicted_score')
plt.title("model prediction vs actual scores")
plt.xlabel("study hours per day")
plt.ylabel("final output")
plt.grid = True
plt.show()
study_hours = float(input("Enter daily study hours: "))
attendance = float(input("Enter attendance percentage: "))
test1 = float(input("Enter Internal Test 1 marks (out of 40): "))
assignment = float(input("Enter Assignment score (out of 10): "))
test2 = float(input("Enter Internal Test 2 marks (out of 40): "))

prediction = model.predict([[study_hours, attendance, test1, assignment, test2]])
print(f"final predicted score is:", prediction)
```
